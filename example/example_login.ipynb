### Importing Logging and SmartConnect

```python
from SmartApi import SmartConnect
from SmartApi.loggerConfig import get_logger 

logger = get_logger(__name__, "INFO")
```

#### Define client info

```python
client_info = {
    "api_key": "",
    "client_id": "",
    "password": "",
    "totp_secret": "",
}
```

### User (Login Services)

##### [SmartAPI User Documentation](https://smartapi.angelbroking.com/docs/User)

#### Generate SmartApi session

```python
import pyotp

try:
    totp = pyotp.TOTP(client_info["totp_secret"]).now()
except Exception as e:
    logger.error("Invalid Token: The provided token is not valid.")
    raise e

api = SmartConnect(api_key=client_info["api_key"])

response  = api.generateSession(client_info["client_id"], client_info["password"], totp)
response
```

#### Get Profile

```python
api.getProfile(refreshToken=api.getrefreshToken)
```

#### Funds and Margins

```python
api.rmsLimit()
```

#### Logout

```python
api.terminateSession(clientCode=client_info["client_id"])
```
