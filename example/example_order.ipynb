### Importing Logging and SmartConnect

```python
from SmartApi import SmartConnect
from SmartApi.loggerConfig import get_logger 

logger = get_logger(__name__, "INFO")
```

#### Define client info

```python
client_info = {
    "api_key": "",
    "client_id": "",
    "password": "",
    "totp_secret": "",
}
```

#### Generate SmartApi session

```python
import pyotp

try:
    totp = pyotp.TOTP(client_info["totp_secret"]).now()
except Exception as e:
    logger.error("Invalid Token: The provided token is not valid.")
    raise e

api = SmartConnect(api_key=client_info["api_key"])

response  = api.generateSession(client_info["client_id"], client_info["password"], totp)
response
```

## Orders

##### [SmartAPI Orders Documentation](https://smartapi.angelbroking.com/docs/Orders)

#### Place Orders

```python
try:
    orderparams = {
        "variety": "NORMAL",
        "tradingsymbol": "SBIN-EQ",
        "symboltoken": "3045",
        "transactiontype": "BUY",
        "exchange": "NSE",
        "ordertype": "LIMIT",
        "producttype": "INTRADAY",
        "duration": "DAY",
        "price": "19500",
        "squareoff": "0",
        "stoploss": "0",
        "quantity": "1"
        }
    
    # Method 1: Place an order and return the order ID
    orderid = api.placeOrder(orderparams)
    logger.info(f"PlaceOrder : {orderid}")
    
    # Method 2: Place an order and return the full response
    response = api.placeOrderFullResponse(orderparams)
    logger.info(f"PlaceOrder : {response}")
    
except Exception as e:
    logger.exception(f"Order placement failed: {e}")
```

#### Modify an existing Order

```python
try:
    modify_order_params = {
        "orderid": orderid,    
        "variety": "NORMAL",
        "tradingsymbol": "SBIN-EQ",
        "symboltoken": "3045",
        "transactiontype": "BUY",
        "exchange": "NSE",
        "ordertype": "LIMIT",
        "producttype": "INTRADAY",
        "duration": "DAY",
        "price": 19600,                 
        "quantity": 2
    }

    modified_order_response = api.modifyOrder(modify_order_params)
    logger.info(f"Order modified successfully: {modified_order_response}")

except Exception as e:
    logger.exception(f"Order modification failed: {e}")
```

#### Cancel Order

```python
try:
    cancel_response = api.cancelOrder(orderid, "NORMAL")
    logger.info(f"Cancel Order Response : {cancel_response}")
except Exception as e:
    logger.exception(f"Order cancellation failed: {e}")
```

#### GTT Create Rule

```python
try:
    gttCreateParams = {
        "tradingsymbol": "SBIN-EQ",
        "symboltoken": "3045",
        "exchange": "NSE",
        "producttype": "MARGIN",
        "transactiontype": "BUY",
        "price": 100000,
        "qty": 10,
        "disclosedqty": 10,
        "triggerprice": 200000,
        "timeperiod": 365
    }
    rule_id = api.gttCreateRule(gttCreateParams)
    logger.info(f"GTT Rule Created: {rule_id}")
except Exception as e:
    logger.exception(f"GTT rule creation failed: {e}")
```

#### GTT Modify Rule

```python
try:
    gttModifyParams = {
        "id": rule_id,
        "symboltoken": "3045",
        "exchange": "NSE",
        "price": 19500,
        "quantity": 10,
        "triggerprice": 200000,
        "disclosedqty": 10,
        "timeperiod": 365
    }
    modified_rule_id = api.gttModifyRule(gttModifyParams)
    logger.info(f"GTT Rule Modified: {modified_rule_id}")
except Exception as e:
    logger.exception(f"GTT rule modification failed: {e}")
```

#### GTT Cancel Rule

```python
try:
    cancelParams = {
        "id": rule_id,
        "symboltoken": "3045",
        "exchange": "NSE"
    }
    cancelled_rule_id = api.gttCancelRule(cancelParams)
    logger.info(f"GTT Rule Cancelled: {cancelled_rule_id}")
except Exception as e:
    logger.exception(f"GTT rule cancellation failed: {e}")
```
