### Importing Logging and SmartConnect

```python
from SmartApi import SmartConnect
from SmartApi.loggerConfig import get_logger 

logger = get_logger(__name__, "INFO")
```

#### Define client info

```python
client_info = {
    "api_key": "",
    "client_id": "",
    "password": "",
    "totp_secret": "",
}
```

#### Generate SmartApi session

```python
import pyotp

try:
    totp = pyotp.TOTP(client_info["totp_secret"]).now()
except Exception as e:
    logger.error("Invalid Token: The provided token is not valid.")
    raise e

api = SmartConnect(api_key=client_info["api_key"])

response  = api.generateSession(client_info["client_id"], client_info["password"], totp)
response
```

### Live Market Data

##### [SmartAPI Market Data Documentation](https://smartapi.angelbroking.com/docs/MarketData)

#### Get LTP Data

```python
try:
    exchange = "NSE" # or "BSE"
    symbol = "SBIN-EQ"
    symbol_token = "3045"
    
    response = api.ltpData(exchange, symbol, symbol_token)
    ltp = response["data"]
    logger.info(f"LTP Data : {ltp}")
except Exception as e:
    logger.exception(f"Failed to fetch LTP data: {e}")
```

#### Get Market Data

```python
try:
    mode = "FULL" # or "OHLC" or "LTP"
    exchangeTokens = {
        "NSE": ["3045"]
    }
    response = api.getMarketData(mode, exchangeTokens)
    marketData = response["data"]["fetched"]
    logger.info(f"Market Data : {marketData}")
except Exception as e:
    logger.exception(f"Failed to fetch market data: {e}")
```

#### Search Scrip

```python
try:
    exchange = "BSE"
    searchscrip = "Titan"
    response = api.searchScrip(exchange, searchscrip)
    searchScripData = response["data"]
    logger.info(f"Search Scrip Data : {searchScripData}")
except Exception as e:
    logger.exception(f"Failed to search scrip: {e}")
```
